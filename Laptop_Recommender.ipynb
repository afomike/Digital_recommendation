```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
import random

# Assuming 'new' is the processed DataFrame and 'similarity' is the cosine similarity matrix
# This function generates predictions by selecting the most similar items based on the similarity matrix
def generate_predictions(similarity_matrix, item_indices, k=10):
    predictions = {}
    for i, item_idx in enumerate(item_indices):
        similar_items = sorted(list(enumerate(similarity_matrix[item_idx])), reverse=True, key=lambda x: x[1])
        recommended_items = [idx for idx, _ in similar_items[:k]]
        predictions[i] = recommended_items
    return predictions

# Select a random subset of item indices for evaluation
num_items_to_evaluate = 600
item_indices_to_evaluate = random.sample(range(len(new)), num_items_to_evaluate)

# Generate predictions for the selected items
predictions = generate_predictions(similarity, item_indices_to_evaluate)

# Evaluate precision and recall
def precision_recall_at_k(predictions, k=10, threshold=3.5):
    num_users = len(predictions)
    precision_sum = 0
    recall_sum = 0

    for user_id in predictions.keys():
        recommended_items = predictions[user_id]
        relevant_items = [idx for idx, _ in enumerate(new['usecases']) if new.iloc[idx]['usecases'] == new.iloc[user_id]['usecases']]
        relevant_items = [item_idx for item_idx in relevant_items if item_idx != user_id]  # Exclude the current item

        num_recommended = len(recommended_items)
        num_relevant = len(relevant_items)
        num_common = len(set(recommended_items).intersection(relevant_items))

        precision_sum += num_common / num_recommended if num_recommended > 0 else 0
        recall_sum += num_common / num_relevant if num_relevant > 0 else 0

    precision_at_k = precision_sum / num_users
    recall_at_k = recall_sum / num_users

    return precision_at_k, recall_at_k

precision_at_10, recall_at_10 = precision_recall_at_k(predictions, k=10, threshold=3.5)
print("Precision@10:", precision_at_10)
print("Recall@10:", recall_at_10)
```
